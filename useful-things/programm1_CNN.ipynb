# CNN Model for Primary Color Recognition


### Introduction

In this notebook, we will build a Convolutional Neural Network (CNN) model to recognize the primary color in a dataset of jpg images. We will use Keras to build the model and OpenCV to process the images.

```python
import sys
print(sys.executable)
import os 
print(os.getcwd())
```

### Import libraries
We will import the required libraries.

```python
import os
import csv
import cv2
import numpy as np
import tensorflow as tf
from keras.models import Sequential
from keras.layers import Conv2D, MaxPooling2D, UpSampling2D, Flatten, Dense
from keras.optimizers import Adam
from sklearn.model_selection import train_test_split
from sklearn.metrics import mean_squared_error, mean_absolute_error
```

### Data Loading
We will load the images and their primary colors from the given paths. The primary color of each image is determined by finding the most common pixel value.

```python
def load_images(image_paths):
    images = []
    for image_path in image_paths:
        image_path = r'{}'.format(image_path)  # Use a raw string literal
        image = cv2.imread(image_path)
        if image is None:
            print(f"Failed to load image at {image_path}")
            continue
        image = cv2.cvtColor(image, cv2.COLOR_BGR2RGB)
        images.append(image)
    return np.array(images)

input_csv_path = 'part1/image_paths_Right.csv'

if not os.path.exists(input_csv_path):
    print(f"Error: {input_csv_path} does not exist")
    exit(1)

with open(input_csv_path, 'r') as input_csv:
    reader = csv.reader(input_csv)
    image_paths = [row[0] for row in reader]

images = load_images(image_paths)
```

### Get Primary Color

```python
def get_primary_color(image):
    pixels = image.reshape(-1, 3)
    primary_color = np.bincount(pixels.astype(int).tostring()).argmax()
    primary_color = np.fromstring(primary_color, dtype=int)
    return primary_color

primary_colors = [get_primary_color(image) for image in images]
```

### Model Buidling Function
We will build a simple CNN model with several convolutional, max pooling, and upsampling layers. The final layer is a convolutional layer with 3 filters and a softmax activation function.

```python
def build_model():
    model = Sequential()
    model.add(Conv2D(64, (3, 3), activation='relu', padding='same', input_shape=(256, 256, 3)))
    model.add(MaxPooling2D((2, 2), padding='same'))
    model.add(Conv2D(128, (3, 3), activation='relu', padding='same'))
    model.add(MaxPooling2D((2, 2), padding='same'))
    model.add(Conv2D(256, (3, 3), activation='relu', padding='same'))
    model.add(UpSampling2D((2, 2)))
    model.add(Conv2D(128, (3, 3), activation='relu', padding='same'))
    model.add(UpSampling2D((2, 2)))
    model.add(Conv2D(3, (3, 3), activation='softmax', padding='same'))
    return model
```

### Model Trainig
We will train the model on our training data for 10 epochs. We will use the Adam optimizer and mean squared error loss function.

model trainig function

```python
def train_model(model, images, primary_colors):
    train_images, val_images, train_colors, val_colors = train_test_split(images, primary_colors, test_size=0.2, random_state=42)
    model.compile(optimizer=Adam(), loss='mse')
    model.fit(train_images, train_colors, epochs=10, validation_data=(val_images, val_colors))
```

### Main Execution
We will read the image paths from a CSV file, load the images and their primary colors, build the model, and train the model.

```python
input_csv_path = 'image_paths_Right.csv'

if not os.path.exists(input_csv_path):
    print(f"Error: {input_csv_path} does not exist")
else:
    with open(input_csv_path, 'r') as input_csv:
        reader = csv.reader(input_csv)
        image_paths = [row[0] for row in reader]

    images, primary_colors = load_images(image_paths)
    model = build_model()
    train_model(model, images, primary_colors)
```

### Calculate Metrics

```python
predictions = model.predict(images)
mse = mean_squared_error(primary_colors, predictions)
mae = mean_absolute_error(primary_colors, predictions)
print(f"Mean Squared Error: {mse}")
print(f"Mean Absolute Error: {mae}")
```
